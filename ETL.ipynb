```python
def cantidad_filmaciones_mes(Mes):
  """
  Se ingresa un mes en idioma EspaÃ±ol. Debe devolver la cantidad de pelÃ­culas que fueron estrenadas
  en el mes consultado en la totalidad del dataset.

  """
  Mes = Mes.lower()
  mes_dict = {
      'enero':1,
      'febrero':2,
      'marzo':3,
      'abril':4,
      'mayo':5,
      'junio':6,
      'julio':7,
      'agosto':8,
      'septiembre':9,
      'octubre':10,
      'noviembre':11,
      'diciembre':12
  }


  # obtenemos el valor numeroo del mes
  mes_num = mes_dict[Mes]
  #Obtenemos la catidad de peliculas que se lanzaron en el mes
  cantidad_filmaciones = data_movies[data_movies['release_date'].dt.month == mes_num].shape[0]
  return str(cantidad_filmaciones) + " es la cantidad de peliculas que fueron estrenadas el mes de "+str(Mes)

```

```python
def cantidad_filmaciones_dia(dia):

    """
    Se ingresa un mes en idioma EspaÃ±ol. Debe devolver la cantidad de pelÃ­culas que fueron estrenadas
    en el dia consultado en la totalidad del dataset.

    """
    dia = dia.lower()
    dia_dict = {
      'uno':1,'dos':2,'tres':3,'cuatro':4,'cinco':5,'seis':6,'siete':7,'ocho':8,'nueve':9,'diez':10,
      'once':11,'doce':12,'trece':13,'catorce':14,'quince':15,'dieciseis':16,'diecisiete':17,'dieciocho':18,
      'diecinueve':19,'veinte':20,'veitiuno':21,'veitidos':22,'veititres':23,'veiticuatro':24,'veiticinco':25,
      'veintiseis':26,'veintisiete':27,'veintiocho':28,'veintinueve':29,'treinta':30,'treinta y uno':31,

  }
    
     # obtenemos el valor numeroo del mes
    dia = dia_dict[dia]
    #Obtenemos la catidad de peliculas que se lanzaron en el mes
    cantidad_filmaciones = data_movies[data_movies['release_date'].dt.month == dia].shape[0]
    return str(cantidad_filmaciones) + " es la cantidad de peliculas que fueron estrenadas el dia "+ str(dia)
```

```python
def votos_titulo( titulo_de_la_filmaciÃ³n ): 

    """Se ingresa el tÃ­tulo de una filmaciÃ³n esperando como respuesta el tÃ­tulo, la cantidad de votos y el valor promedio 
    de las votaciones. La misma variable deberÃ¡ de contar con al menos 2000 valoraciones, caso contrario, se regresa un mensaje avisando que no cumple
      esta condiciÃ³n y que por ende, no se devuelve ningun valor."""
    
    titulos_pelis = data_movies["title"].unique()

    if titulo_de_la_filmaciÃ³n  in titulos_pelis:

        vote_count = data_movies.loc[data_movies["title"] == titulo_de_la_filmaciÃ³n ,"vote_count"]
        vote_average = data_movies.loc[data_movies["title"] == titulo_de_la_filmaciÃ³n ,"vote_average"]
        estreno = data_movies.loc[data_movies["title"] == titulo_de_la_filmaciÃ³n ,"release_year"]
        
        estreno = int(estreno.iloc[0])
        vote_count = float(vote_count.iloc[0])
        vote_average = float(vote_average.iloc[0])
        if vote_count >= 2000:
        
            return "La pelicula " + titulo_de_la_filmaciÃ³n + " fue estrenada en el aÃ±o "+ str(estreno) + " , la misma cuenta con un total de  "+ str(vote_count)+" valoraciones y con un promedio de "+str(vote_average)
        else :
            print("Esta pelicula no cuenta con la valoracion necesaria no a sido de agrado para el publico")

        
    else:
        return print("Esta pelicula no esta disponible dentrode nuestro catalogo")

        #raise HTTPException(status_code=400, detail="Error al ingresar el nombre de la pelicula o no se encuntra dentro de lista de peliculas de eta plataforma")

    

    
```

```python
def get_actor( nombre_actor ):
    """ Se ingresa el nombre de un actor que se encuentre dentro de un dataset debiendo devolver el Ã©xito del mismo medido a travÃ©s del retorno. 
    AdemÃ¡s, la cantidad de pelÃ­culas que en las que ha participado y el promedio de retorno."""

    val_return = []
    count = 0

    for index,row in data_credits.iterrows():
        
            lista_actors = row["Actors/Actress"]
            

            if nombre_actor in lista_actors:
                retun = row["return"]
                val_return.append(retun)
                count +=1

    suma_return = round(sum(val_return) ,3) 
    promedio = round(suma_return / len(val_return) ,3)     

    return "El actor " + nombre_actor + " ha participado en " + str(count) + " peliculas , el mismo ha consegido un retorno de "+str(suma_return)+" con un promedio de "+str(promedio)
                
    
```

```python
def get_director( nombre_director ): 
    """Se ingresa el nombre de un director que se encuentre dentro de un dataset debiendo devolver el Ã©xito del mismo medido a travÃ©s del retorno.
      AdemÃ¡s, deberÃ¡ devolver el nombre de cada pelÃ­cula con la fecha de lanzamiento, retorno individual, costo y ganancia de la misma"""
    
    val_return = []
    count = 0
    name = []
    fecha = []
    costo = []
    ganancia = []

  

    for index,row in data_credits.iterrows():
        
            lista_director = row["Director"]
            

            if nombre_director in lista_director:
                retun = row["return"]
                name.append(row["title"])
                fecha.append(row["release_year"])
                costo.append(row["budget"])
                ganancia.append(row["revenue"])
                val_return.append(retun)
                count +=1

    suma_return = round(sum(val_return) ,3) 
    

    
    lista_dict = []
    
    
    # Iterar sobre las claves y los valores combinados
    for name,fecha,costo,ganancia in zip(name,fecha,costo,ganancia):
      dicc = {
           "Director":nombre_director,
           "Retorno":suma_return,
           "Numero pelis":count,
           "Titulo Pelicula":name,
           "Fecha lanzamiento":fecha,
           "Costo":costo,
           "Ganancia":ganancia
         
            }
      lista_dict.append(dicc)

    return    lista_dict
    
```
